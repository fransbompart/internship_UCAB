```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv('data-clickup.csv')

# limpiar columnas
dataset.drop(['Assignee', 'Status', 'Time Estimate', 'Time Logged', 'Task ID'], axis=1, inplace=True)

# correlacion

correlation_matrix = dataset.corr()
correlation_matrix['Actual Hours'].sort_values(ascending=False)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
